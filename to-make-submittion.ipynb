```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

#30000万人に対しての予測結果の読み取り
result_30000=pd.read_csv("../input/submit/preds_30000.csv")

#残りの顧客に対しての予測結果の読み取り
result_remain=pd.read_csv("../input/submit/preds_remain.csv")

#sample_submittionの読み取り
SS=pd.read_csv("../input/rakusp/for_uploading/sample_submission_rakus_latest.csv")
```

```python
#30000万人に対しての予測結果と残りの顧客に対しての予測結果を結合
all_submittion = pd.concat([result_30000, result_remain])
```

```python
#予測結果の表示
all_submittion
```

```python
#sample_submittionのprediction列の削除
del SS["prediction"]
SS
```

```python
#sample_submittionと最終予測結果の結合
submittion = pd.merge(SS, all_submittion, on='customer_id', how='inner')
```

```python
#提出ファイル完成
submittion

#ファイルの書き出し
submittion.to_csv('/kaggle/working/submission.csv', index=False)
```
